```python
import matplotlib.pyplot as plt
import random

# Generate 100 random x and y coordinates
x_coordinates = [random.uniform(0, 10) for _ in range(100)]
y_coordinates = [random.uniform(0, 3) for _ in range(100)]

# Plot the points
plt.scatter(x_coordinates, y_coordinates)

# Set labels for x and y axes
plt.xlabel('X')
plt.ylabel('Y')

# Set the title for the plot
plt.title('100 random dots')

# Show the plot
plt.show()
```
